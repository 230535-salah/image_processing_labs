```python
# gamma correction

import cv2
import numpy as np

image_path = "assets/sample_image_02.png"
output_path = "assets/output.png"

gamma = 0.6
c = 1.0

image = cv2.imread(image_path)
if image is None:
    raise ValueError("Image not found")

image_np = np.array(image)

image_norm = image_np.astype("float32") / 255.0
image_gamma = c * (image_norm ** gamma)

image_gamma = np.clip(image_gamma, 0.0, 1.0)
image_out = (image_gamma * 255.0).astype("uint8")

cv2.imwrite(output_path, image_out)

```

```python
import cv2
import numpy as np
import os

def contrast_stretching(image, s_min, s_max):
    image = image.astype("float32")

    r_min = np.min(image)
    r_max = np.max(image)

    stretched = ((s_max - s_min) / (r_max - r_min)) * (image - r_min) + s_min

    return stretched
```

```python
input_dir = "lab_images"
output_dir = "lab_images_output"
os.makedirs(output_dir, exist_ok=True)

for filename in os.listdir(input_dir):
    path = os.path.join(input_dir, filename)
    image = cv2.imread(path)

    if image is None:
        continue

    result = contrast_stretching(image, 0, 255)

    result = np.clip(result, 0, 255).astype("uint8")

    cv2.imwrite(os.path.join(output_dir, filename), result)
```

```python
output_dir = "lab_images_output_normalized"
os.makedirs(output_dir_norm, exist_ok=True)

for filename in os.listdir(input_dir):
    path = os.path.join(input_dir, filename)
    image = cv2.imread(path)

    if image is None:
        continue

    image_norm = image.astype("float32") / 255.0

    result = contrast_stretching(image_norm, 0, 1)

    result = np.clip(result, 0, 1)
    result = (result * 255.0).astype("uint8")

    cv2.imwrite(os.path.join(output_dir_norm, filename), result)
```

```python
image = cv2.imread("lab_images/1.png")

result1 = contrast_stretching(image, 50, 255)
result2 = contrast_stretching(image, 100, 160)

result1 = np.clip(result1, 0, 255).astype("uint8")
result2 = np.clip(result2, 0, 255).astype("uint8")

cv2.imwrite("lab_images_result/result_50_255.jpg", result1)
cv2.imwrite("lab_images_result/result_100_160.jpg", result2)
```
